# Les données:
Le dossier "data/raw" contient des milliers d'enregistrements audio de 10 personnes (5 hommes et 5 femmes) américaines. Le but de ce projet et d'analyser ces données et de construire un modèle simple de reconnaissance vocale. C'est à dire, reconnaitre la personne qui parle parmi les 10. Voici à quoi ressemble un enregistrement:

```python
# pip install librosa
```

```python
from IPython.display import Audio
import librosa

y1, sr1 = librosa.load("data/raw/f0001_us_f0001_00001.wav", sr=None)
Audio(data=y1, rate=sr1)

```

```python
print(y1)
print(f"La taille du fichier audio est {y1.shape}")
```

Nous allons commencer par extraire des informations sur le timbre (en utilisant le domaine temps-fréquence) de l'enregistrement avec les Mel Frequency Cepstral Coefficient (MFCC):

```python
mfcc = librosa.feature.mfcc(y=y1, sr=sr1, n_mfcc=10)
print(mfcc)
print(f"La taille des coefficients est {mfcc.shape}")
```

On peut visualiser ces 10 coefficients dans le temps:

```python
from matplotlib import pyplot as plt

# Plotting the MFCCs
plt.figure(figsize=(4, 3))
librosa.display.specshow(mfcc, x_axis='time', sr=sr1)
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')
plt.tight_layout()
plt.show()
```

Nous allons résumer ces coefficients en calculant leur moyenne temporelle. Ainsi, chaque fichier audio doit être transformé par un vecteur de taille 13:

```python
mfcc_moyenne = mfcc.mean(axis=1)
mfcc_moyenne
```

On construit donc le dataset avec la fonction suivante:


```python
import pandas as pd

folder_path = "data/raw/"
n_recordings = 50
n_coefs = 10
data = []
for gender in ["m", "f"]:
    for speaker in range(1, 6):
        for recording_id in range(1, n_recordings + 1):
            file_name = "%s%04d_us_%s%04d_%05d.wav" %(gender, speaker, gender, speaker, recording_id)
            y, sr = librosa.load(folder_path + file_name)
            mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_coefs).mean(axis=1)

            speaker_id = speaker - 1
            if gender == "f":
                speaker_id += 5
            row_data = {
                'FileName': file_name,
                'Gender': gender,
                'SpeakerID': speaker_id,
                'RecordingID': recording_id,
            }
            for i in range(n_coefs):
                row_data[f'MFCC_{i+1}'] = mfcc[i]
            
            data.append(row_data)

df = pd.DataFrame(data)
```

**- Hypothèse nulle $ H_0 \$ :** Les enregistrements pairs et impairs n'ont pas de différences significatives sur les dimensions Dim1 et Dim2. Cela implique que la moyenne multivariée des enregistrements pairs est la même que celle des enregistrements impairs :  

$$ \mu_{\text{pair}}  =  \mu_{\text{impair}} $$

où $ \mu_{\text{pair}} $ et $ \mu_{\text{impair}} $ représentent respectivement les moyennes multivariées des enregistrements pairs et impairs.

**- Hypothèse alternative $ H_1 $ :** Il existe une différence significative entre les enregistrements pairs et impairs sur au moins une des dimensions :

$$ \mu_{\text{pair}}  ≠  \mu_{\text{impair}} $$

```python
import pandas as pd
from scipy.stats import ttest_ind, mannwhitneyu
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Charger les données
# Suppose que MFCC_pca_df est déjà chargé comme DataFrame

# 2. Filtrer les enregistrements de "Speaker 4"
speaker_4_df = MFCC_pca_df[MFCC_pca_df["SpeakerID"] == 4]

# Séparer les données en deux groupes : enregistrements pairs et impairs
group_even = speaker_4_df[speaker_4_df["RecordingID"] % 2 == 0]
group_odd = speaker_4_df[speaker_4_df["RecordingID"] % 2 != 0]

# 3. Visualisation des distributions de Dim1 et Dim2 pour les groupes pairs et impairs
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=[group_even["Dim1"], group_odd["Dim1"]], palette="Set2")
plt.xticks([0, 1], ["Even RecordingID", "Odd RecordingID"])
plt.title("Distribution of Dim1 for Even and Odd RecordingIDs")

plt.subplot(1, 2, 2)
sns.boxplot(data=[group_even["Dim2"], group_odd["Dim2"]], palette="Set2")
plt.xticks([0, 1], ["Even RecordingID", "Odd RecordingID"])
plt.title("Distribution of Dim2 for Even and Odd RecordingIDs")
plt.show()

# 4. Tester la normalité des données pour choisir le test approprié
# Utilisation de tests de Shapiro-Wilk ou d'une observation visuelle des histogrammes (omise ici)

# 5. Tests statistiques pour Dim1 et Dim2 entre groupes pairs et impairs
# Test t de Student
t_stat_dim1, p_value_dim1 = ttest_ind(group_even["Dim1"], group_odd["Dim1"])
t_stat_dim2, p_value_dim2 = ttest_ind(group_even["Dim2"], group_odd["Dim2"])

# Test de Mann-Whitney U (au cas où les distributions ne sont pas normales)
u_stat_dim1, p_value_u_dim1 = mannwhitneyu(group_even["Dim1"], group_odd["Dim1"])
u_stat_dim2, p_value_u_dim2 = mannwhitneyu(group_even["Dim2"], group_odd["Dim2"])

# 6. Résultats
print("Résultats du test t de Student:")
print(f"Dim1: t-stat = {t_stat_dim1}, p-value = {p_value_dim1}")
print(f"Dim2: t-stat = {t_stat_dim2}, p-value = {p_value_dim2}")

print("\nRésultats du test de Mann-Whitney U:")
print(f"Dim1: U-stat = {u_stat_dim1}, p-value = {p_value_u_dim1}")
print(f"Dim2: U-stat = {u_stat_dim2}, p-value = {p_value_u_dim2}")

# Interprétation des résultats :
if p_value_dim1 < 0.05:
    print("Il y a une différence statistiquement significative entre les groupes pairs et impairs pour Dim1 (test t de Student).")
else:
    print("Aucune différence significative pour Dim1 entre les groupes pairs et impairs (test t de Student).")

if p_value_dim2 < 0.05:
    print("Il y a une différence statistiquement significative entre les groupes pairs et impairs pour Dim2 (test t de Student).")
else:
    print("Aucune différence significative pour Dim2 entre les groupes pairs et impairs (test t de Student).")

if p_value_u_dim1 < 0.05:
    print("Il y a une différence statistiquement significative entre les groupes pairs et impairs pour Dim1 (test de Mann-Whitney U).")
else:
    print("Aucune différence significative pour Dim1 entre les groupes pairs et impairs (test de Mann-Whitney U).")

if p_value_u_dim2 < 0.05:
    print("Il y a une différence statistiquement significative entre les groupes pairs et impairs pour Dim2 (test de Mann-Whitney U).")
else:
    print("Aucune différence significative pour Dim2 entre les groupes pairs et impairs (test de Mann-Whitney U).")
```

**- Test de Hotelling :**

```python
speaker4_pair = speaker_4_df_pair[['Dim1', 'Dim2']]
speaker4_impair = speaker_4_df_impair[['Dim1', 'Dim2']]
```

```python
# Taille des échantillons
n1, n2 = len(speaker4_pair), len(speaker4_impair)
```

```python
# Calcul des moyennes des deux groupes
mean1 = speaker4_pair.mean().values
mean2 = speaker4_impair.mean().values
```

```python
# Calcul des matrices de covariance des deux groupes
cov1 = np.cov(speaker4_pair.T)
cov2 = np.cov(speaker4_impair.T)
```

```python
# Matrice de covariance pondérée (estimée)
Sp = ((n1 - 1) * cov1 + (n2 - 1) * cov2) / (n1 + n2 - 2)
```

```python
# Différence entre les moyennes
mean_diff = mean1 - mean2

# Calcul de la statistique T^2
T_squared = (n1 * n2) / (n1 + n2) * mean_diff.T @ np.linalg.inv(Sp) @ mean_diff
```

```python
# Conversion de T^2 en statistique F
p = len(mean_diff)  # Nombre de dimensions (2 dans ce cas)
F_statistic = (T_squared * (n1 + n2 - p - 1)) / ((n1 + n2 - 2) * p)

# Degrés de liberté pour le test F
df1, df2 = p, n1 + n2 - p - 1

# Calcul de la p-valeur
p_value = 1 - f.cdf(F_statistic, df1, df2)
```

```python
# Affichage des résultats
print("Statistique T^2 de Hotelling:", T_squared)
print("Statistique F correspondante:", F_statistic)
print("Degrés de liberté:", df1, "et", df2)
print("p-valeur:", p_value)
```

```python
# Interprétation du résultat
alpha = 0.05  # Niveau de signification
if p_value < alpha:
    print("Il y a une différence significative entre les deux groupes.")
else:
    print("Aucune différence significative n'a été trouvée entre les deux groupes.")
```
